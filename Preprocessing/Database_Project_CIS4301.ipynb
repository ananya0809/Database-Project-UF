```python
data = pd.read_csv("crime_2010_2019.csv")
```

```python
data
```

```python
data.loc[:,["Date Rptd", "DATE OCC"]]
```

```python
date_of_report = data["Date Rptd"].values.tolist()
date_of_crime = data["DATE OCC"].values.tolist()
```

```python
date_of_report
```

```python
date_of_crime
```

```python
report = []
crime = []
for i in range(len(date_of_report)):
  report.append(str(date_of_report[i]))
for i in range(len(date_of_crime)):
  crime.append(str(date_of_crime[i]))
```

```python
report
```

```python
crime
```

```python
for i in range(len(report)):
  report[i] = report[i][-16:-12] + "-" + report[i][:2] + "-" + report[i][3:5]
report
```

```python
for i in range(len(crime)):
  crime[i] = crime[i][-16:-12] + "-" + crime[i][:2] + "-" + crime[i][3:5]
crime
```

```python
len(report)
```

```python
len(crime)
```

```python
data.drop("Date Rptd", axis = 1, inplace = True)
data["Date Rptd"] = report
data.drop("DATE OCC", axis = 1, inplace = True)
data["Date Occ"] = crime
```

```python
data
```

```python
data = data.drop(['Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street', 'Weapon Used Cd'], axis=1)
```

```python
data['Weapon Desc'] = data['Weapon Desc'].fillna('No Entry')
```

```python
data
```

```python
data['Date Rptd'] = pd.to_datetime(data['Date Rptd'])
prev_years = data[~(data['Date Rptd'] > '2016-12-31')]
prev_years
```

```python
modf_data = pd.merge(data,prev_years, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)
```

```python
modf_data.to_csv('crimedataset2017to2022.csv')
```

```python
modf_data = pd.read_csv("crimedataset2017to2022.csv")
modf_data
```

```python
modf_data.loc[:,"TIME OCC"]
```

```python
time_of_crime = modf_data["TIME OCC"].values.tolist()
```

```python
time_of_crime
```

```python
len(time_of_crime)
```

```python
time_of_crime[0]
```

```python
type(time_of_crime[0])
```

```python
crimetime = []
for i in range(len(time_of_crime)):
  crimetime.append(str(time_of_crime[i]))
crimetime
```

```python
for i in range(len(crimetime)):
  if len(str(crimetime[i])) > 1:
    crimetime[i] = crimetime[i][:-2] + ":" + crimetime[i][-2:] + ":" + "00"
  else:
    crimetime[i] = crimetime[i][:] + ":" + "00" + ":" + "00"
crimetime
```

```python
for i in range(len(crimetime)):
  if len(crimetime[i][:-7]) != 1:
    crimetime[i] = "0" + crimetime[i][:]
crimetime
```

```python
modf_data.drop("TIME OCC", axis = 1, inplace = True)
modf_data["Time Occ"] = crimetime
```
